### def sortGff3

```python
def sortGff3(gff3_File, types = ['gene','mRNA','exon']): ## don't forget to add name checking
    """   
    This checks for proper "structure" for correct format:
    - file starts with proper header
    - each line is tab delimited and has 9 components.
    - each line has a proper type at the 3rd component.
    - example line:
    1stComp\t2ndComp\t3rdComp\t4thComp\t...
    
    It also sorts each line in the gff file according to the location of the entry and the
    type hierarchy. Valid types are 'gene', 'mRNA', 'exon' and must match exactly. 
    Gene != gene, mrna != mRNA, etc.
    
    Parameters:
    -'gff3_File': the uploaded gff file.
    -'types': a list indicating the types of each gff line and the order the types should
              be sorted in. The first type in the list will be ordered before the second, 
              the second type before the third, etc.

    Output:
    -'keyList': a list of keys sorted in the order that the lines will be sorted.
    -'holder': a dictionary of lines paired with keys in 'keyList'.
    """
    
    holder = dict()                    #keys are <type>_leftcoord; value is the entire line
    counter = dict()                   #keys are leftcoord; values are sum of indexs of types found with types list
    structure = []                     #this will hold the errors from this function, which are considered "structural"
    contigCount = False

    formatCount = 0
    
    with open(gff3_File, "r") as file:
        fi = file.readlines()              #OK for phages, may fail for large gff files

    try:   #check if first line is proper header
        if fi[0] != "##gff-version 3\n":
            raise LineError("0026")
    except LineError as er:
        structure.append("[ 0 ] " + er.returnError())

        
    for lineCount,line in enumerate(fi[1:]):

        try:
            theLine = line.strip().split("\t")  #split on tab
            if len(theLine) > 9: # 9 columns in a proper gff file
                raise FormatError("0200")
            elif len(theLine) < 9:
                raise FormatError("0100")
        except FormatError as er:
            structure.append("[line# " + str(lineCount+2) + "] " + er.returnError())
            continue
        except:
            structure.append("Validation Error: unkown error. Check line# " + str(lineCount+2) + ":\n" + line)

        if theLine[2] == "contig":
            holder[theLine[2] + "_" + theLine[3]] = line
            continue

        try:
            if theLine[2] not in types:
                raise FormatError("0700")
        except FormatError as er:
            structure.append("[" + str(lineCount+2) + "] " + er.returnError())
            continue

        try:
            leftPosition = int(theLine[3])
            rightPosition = int(theLine[4])
        except ValueError as er:
            structure.append("[" + str(lineCount+2) + "] " + "coordinates not readable as numbers in column 4 and/or 5")
            continue

        #to get here line will is partially validated, it will:
        # 1. not be type contig
        # 2. have 9 components 
        # 3. have a valid type in 3rd column
        # 4. have integers in 4th and 5th column

        holder[theLine[2] + "_" + theLine[3]] = line #FIXME: problem if 2 genes with same start number (chaperones)

        if theLine[3] in counter:
            counter[theLine[3]] = counter[theLine[3]] + types.index(theLine[2])
        else:
            counter[theLine[3]] = types.index(theLine[2])


    #this part checks that you have the correct number of entries for each start coordinate:
    counterKeys = counter.keys()
    
    expectedNum = ((len(types))*(len(types)-1))/2
    print "coutnerKeys are: " + str(counterKeys)

    for key in counterKeys:  
        thisNum = counter[key]
        try:
            if thisNum == expectedNum:
                continue
            else:
                raise FormatError("0500")
        except FormatError as er:
            structure.append("Gene at Coordinate " + key + " " + er.returnError())

    Errors["structure"] = structure                    #add the list of structural errors to the Errors dictionary
    
    #need to sort all the keys 
    keyList = holder.keys()
    

    #first sort on the secondary charachteriestic which is type:
    typeHierarchy = ['contig','gene','mRNA','exon']
    typeSortList = sorted(keyList, key=lambda item: typeHierarchy.index(item.split("_")[0] ))
                      
    #now sort on the primary charachteristic which is the left coordinate 
    sortedKeyList = sorted(typeSortList, key=lambda item: int(item.split("_")[1]))
            
    tempS = str(gff3_File)
    newName = tempS[0:-5] + "_sorted.gff3"
    with open(newName,"w") as outfile:
        outfile.write("##gff-version 3\n")
        for key in sortedKeyList:
            outfile.write(holder[key])
            
    return [sortedKeyList, holder]
```

### def fileCheck

```python
def fileCheck(keyList, holder, Seq, types = ['gene','mRNA','exon']):
    """
    Checks each component of each line of the gff file for proper format. These are considered "entry" errors.
    All entry errors will be added to Errors dictionary.

    Parameters:
    -'keyList': list of sorted keys.
    -'holder': dictionary of lines paired with keys in 'keyList'.
    -'Seq': string of the nucleotide sequence extracted from uploaded fasta file.
    -'types': a list indicating the types of each gff line and the order the types should
              be sorted in. The first type in the list will be ordered before the second, 
              the second type before the third, etc.

"""
        
    entry = []                  
    idList = set()                 #list of ID's seen to check for uniquenss
    
    
    
    for line,key in enumerate(keyList[1:]):   #skipping first line as it is the contig
        attributes = dict()                   #will be used to hold each Key/value pair of column 9 for each line
        
        lineCount = line + 1    #adding 1 for unacccounted for header line
        
        try:
            theLine = holder[key].strip().split("\t") ## try here to catch if  not tab deliminated or adding extra lines
        except:
            entry.append("Format Error: each line needs to be tab deliminated.")
            continue
        
        
        if len(theLine) != 9:   #this is just to skip improper lines, error already caught by sortGFF3
            continue
        
            
        ### 1ST and 2ND ITEM ### 
        try:
            if charCheck(theLine[0]):
                raise LineError("0002")
            elif charCheck(theLine[1]):
                raise LineError("0003")
        except LineError as er:
            entry.append("[" + str(lineCount) +"] " + er.returnError())
        
        ### 3RD ITEM ###
        try:
            if theLine[2] not in types:
                raise LineError("0004")  ## types can be changed if more types of line needed
        except LineError as er:
            entry.append("[" + str(lineCount) +"] " + er.returnError())
            continue                     ##no other checks done if not a proper type
        
        ### 4TH ITEM ###
        try:
            if not int(theLine[3]) > 0:
                raise LineError("0005") ## needs to be greater than 0
        except LineError as er:
            entry.append("[" + str(lineCount) +"] " + er.returnError())
        except ValueError:
            entry.append("[" + str(lineCount) + "] coordinates in column 4 not readable as number") 
        
        
        ### 5TH ITEM ###
        try:
            if not int(theLine[4]) > int(theLine[3]):
                raise LineError("0006") ## needs to be greater than 1st coordinate
        except LineError as er:
            entry.append("[" + str(lineCount) +"] " + er.returnError())
        except ValueError:
            entry.append("[" + str(lineCount) + "] coordinates in column 5 not readable as number") 
        
        ### 6TH ITEM ###
        try:
            if theLine[5] != ".": ## can also be a number
                try:
                    int(theLine[5])
                except ValueError:
                    raise LineError("0007")
        except LineError as er:
            entry.append("[" + str(lineCount) +"] " + er.returnError())
            
        ### 7TH ITEM ###
        try:
            if theLine[6] != "+":
                if theLine[6] != "-":
                    raise LineError("0008")  ## + or - strand
        except LineError as er:
            entry.append("[" + str(lineCount) +"] " + er.returnError())
                
         ### 8TH ITEM ### 
        try:
            if theLine[7] != ".":
                raise LineError("0009")
        except LineError as er:
            entry.append("[" + str(lineCount) +"] " + er.returnError())
                
        
        ### 9TH ITEM ###      
        ### 9th item is an attributes list ";" separated list of key/value pairs in the form key=value
       
        #start by checking for invalid characters
        try:
            if charCheck(theLine[8]):
                raise LineError("0011")
        except LineError as er:
            entry.append("[" + str(lineCount) +"] " + er.returnError())
            continue              #don't even try to parse if invalid charachters, just move to next line 


        #checking for proper separation of key/value pairs with the ';' 
        try:
            if theLine[8].count("=") > (theLine[8].count(";") + 1):
                raise LineError("0028")
        except LineError as er:
            entry.append("[" + str(lineCount) +"] " + er.returnError())
            continue               #again don't even try with this much failure
                
        #ok go ahead and split into the underlyine key=value items    
        attrItems = theLine[8].split(";")   #list of each attribute pair

   
        #run through the list of attrItems; make you they are in key=value and put everything in attributes dict
        for attr in attrItems:
            if len(attr) == 0:           #catch attr from extra ; but just skip it
                continue
            try:        ## start with "=" search
                if attr.find("=") < 0:
                    raise LineError("0027")
            except LineError as er:
                entry.append("[" + str(lineCount) +"] " + er.returnError())
                continue                 #if it doesn't have an "=" don't bother with the checks below
                    
            #to get to here attr has to have an "="; so split to check each of key and value
            attributeName, attributeValue = attr.split("=")
            try:                        #make sure of both a attributeName (key) and attributeValue (value)
                if len(attributeName) < 1 or len(attributeValue) < 1:
                    raise LineError("0027")
                else:                   #ok to add to attributes dict
                    attributes[attributeName] = attributeValue
            except LineError as er:
                entry.append("[" + str(lineCount) +"] " + er.returnError())
                    
        #start parsing the attributes dictionary
        try:            #do checks for ID presence and uniqueness
            if "ID" in attributes:
                try:
                    if attributes["ID"] in idList:
                        raise LineError("0012") ## each id can only be used once 
                    else:
                        idList.add(attributes["ID"])
                except LineError as er:
                        entry.append("[" + str(lineCount) +"] " + er.returnError())
                                
            else:
                raise LineError("0013")  ##ID is missing
        except LineError as er:
            entry.append("[" + str(lineCount) +"] " + er.returnError())

                     

        ######################## checks that are specific to certain types #############
        if theLine[2] == types[0]:       #is a gene line; check the biology and check for a Name
            #geneCheck(int(theLine[3]),int(theLine[4]),Seq,lineCount)
            try:  #do checks for Name presence
                if "Name" in attributes:
                    _Name = True
                else:
                    raise LineError("0013")
            except LineError as er:
                entry.append("[" + str(lineCount) +"] " + er.returnError())

        elif types.index(theLine[2]) > 0:    #is an mRNA or exon, check for parent
            try:  #do checks for parent presence
                if "Parent" in attributes:
                    _Parent = True
                else:
                    raise LineError("0025")
            except LineError as er:
                entry.append("[" + str(lineCount) +"] " + er.returnError())
        
        else:
            continue                   #this should not happen, but just in case, skip to the next line
        
                
    Errors["entry"] = entry                    #add the list of entry errors to the Errors dictionary

    return "clean"
    
```

### def charCheck

```python
def charCheck(str, search=re.compile(r'[^a-zA-Z0-9.=;_]').search):
    """
    Checks a string for characters not a-zA-Z0-9.=; and returns true if such a character is
    found.

    Parameters:
    -'str': string being checked.
    """
    return bool(search(str))
```

### def geneCheck

```python
def geneCheck(coord1, coord2, seq, count):       
    """
    Checks if coordinates of sequence given is a proper gene. Writes errors to global list
    'Errors'.
    
    Proper gene:
    - can be translated to a protein.
    - has a stop codon.
    - no internal stop codon.
    - begins with a start codon.
    
    Parameters:
    -'coord1': start coordinate of gene.
    -'coord2': second coordinate of gene.
    -'Seq': string nucleotide sequence.
    -'count': line in the file the gene is from.
    """
    #FIXME: need to add check that all coord are within the range of the contig
    biology = dict()                     #dictionary of errors based on biology like start codon, stop codon
    gene = seq[coord1-1:coord2]
    try:
        if len(gene)%3 != 0:
            raise BiologyError("0050") ## divisible by three
    except BiologyError as er:
        biology.append("[" + str(count) +"] " + er.returnError())
    
    try:
        protein = translate(gene) ## errors in translating the gene
    except:
        biology.append("[" + str(count) + "]  Biology Error: Ensure sequence provided is a nucleotide sequence.")
        return

    Start = True
    startCodon = seq[coord1-1:coord1+2]
    if startCodon == "ATG" or startCodon == "TTG" or startCodon == "GTG":
        Start = False
    
    try:
        if protein.count("*") == 0:
            raise BiologyError("0030") ## no stop codon
        if protein.count("*") > 1:
            raise BiologyError("0040") ## internal stop codons
        if Start:
            raise BiologyError("0020") ## has to start with M
    except BiologyError as er:
        biology.append("[" + str(count) +"] " + er.returnError())
    
    Errors["biology"] = biology                    #add the list of biology errors to the Errors dictionary
    return
    
```

### def fastaRead

```python
def fastaRead(fasta_File):
    """
    Reads in a nucleotide sequence from a text file in .fasta format.

    Parameters:
    -'fasta_File': uploaded text file in .fasta format

    Output:
    -'seq': string of a nucleotide sequence.
    """

    seq = []
    try:
        f1 = open(fasta_File, "r")
    except:
        return "Error: unable to read fasta file."
    
    for line in f1:
        line = line.rstrip()
        if line.startswith(">"):
            continue
        else:
            seq.append(line)

    f1.close()
    return ''.join(seq)
```

### def translate

```python
def translate(nucSeq):
    """
    Translates a nucleotide sequence into a protein sequence. Uses a dictionary of nucleotides
    paired with the protein they translate for.

    Parameters:
    -'nucSeq': string of nucleotides.

    Output:
    -'protSeq': string of proteins.
    """
    
    codonLib = {'TTT':'F','TTC':'F','TTA':'L','TTG':'L','CTT':'L','CTC':'L','CTA':'L',
                'CTG':'L','ATT':'I','ATC':'I','ATA':'I','ATG':'M','GTT':'V','GTC':'V',
                'GTA':'V','GTG':'V','TCT':'S','TCC':'S','TCA':'S','TCG':'S','CCT':'P',
                'CCC':'P','CCA':'P','CCG':'P','ACT':'T','ACC':'T','ACA':'T','ACG':'T',
                'GCT':'A','GCC':'A','GCA':'A','GCG':'A','TAT':'Y','TAC':'Y','TAA':'*',
                'TAG':'*','CAT':'H','CAC':'H','CAA':'Q','CAG':'Q','AAT':'N','AAC':'N',
                'AAA':'K','AAG':'K','GAT':'D','GAC':'D','GAA':'E','GAG':'E','TGT':'C',
                'TGC':'C','TGA':'*','TGG':'W','CGT':'R','CGC':'R','CGA':'R','CGG':'R',
                'AGT':'S','AGC':'S','AGA':'R','AGG':'R','GGT':'G','GGC':'G','GGA':'G','GGG':'G'}
    
    cntLoc = 0
    cntCodon = 0
    codon = ''
    protSeq = ''
    while cntLoc < len(nucSeq):
        codon = codon + nucSeq[cntLoc]
        cntCodon += 1
        if cntCodon == 3:
            temp = codonLib[codon]
            protSeq = protSeq + temp
            codon = ''
            cntCodon = 0
        cntLoc += 1
        
    return protSeq
```

### def outFile

```python
def outFile(keyS, holderS, nameS, errors, nameE, incLine):
    """
    Writes a sorted gff file and an errors text file.
    
    Parameters:
    -'keyS': list of sorted keys.
    -'holderS': dictionary of lines from the gff file paired with keys in 'keyS'.
    -'nameS': name of new sorted gff file. Also location where the file is written.
    -'errors': list of all errors found in uploaded files to be written to text file.
    -'nameE': name of new errors text file. Also location where the file is written.
    -'incLine': boolean indicating whether lines from the sorted file should be included in 
                errors file.
    """
    outE = nameE
    outS = nameS
   
    if incLine:
       for item in errors:
          if "Coordinate" in item:
             tempE = item.split(" ")[1]
             for key in keyS:
                if tempE == key.split("_")[1]:
                   outE.write(holderS[key])
                   break
          else:
             tempE = int(item[1])
             outE.write(holderS[keyS[tempE]])
         
          outE.write(item)
          outE.write("\n")
          outE.write("\n")
    else:
       for item in errors:
          outE.write(item)
          outE.write("\n")
   
    for key in keyS:
       outS.write(holderS[key])
       outS.write("\n")
```

## Define Error Classes

```python
class ValidationError(Exception):
    """
    Exception classes:
    - 'Validation Error': basic exception class. Used to catch unknown errors. 
    - 'Format Error': catches errors dealing with line format.
    - 'Line Error': catches errors dealing with individuel components of a line.
    - 'Biology Error': catches errors dealing with the described genes.
    """
    def __init__(self, code, message = ""):
        
        _dict = dict()
        _dict['0000'] = "Usage Error: Unknown error code given used. Use printDict() to see dictionary of errors"
        _dict['1000'] = "Validation Error: unknown"
        
        self.code = str(code)
        self._dict = _dict
        self.message = message
        
        super(ValidationError, self).__init__(code, message)
        
    def printDict(self):
        for k, v in self._dict.iteritems():
            print k + ":  " + v
            
    def returnError(self):
        try:
            return self._dict[self.code]
        except:
            return self._dict['0000']

class FormatError(ValidationError):
    def __init__(self, code, message = ""):
        
        super(FormatError, self).__init__(code, message)
        
        self._dict['0100'] = "Format Error: less than 9 columns, no other checks done."
        self._dict['0200'] = "Format Error: more than 9 columns, no other checks done."
        self._dict['0300'] = "Format Error: tab deliminated."
        self._dict['0400'] = "Format Error: multiple contigs. There should only be one contig line per file."
        self._dict['0500'] = "Format Error: Errors in, or improper number of types for one location. Default types are [gene, mRNA, exon]"
        self._dict['0600'] = "Format Error: ensure only one sequence in fasta file."
        self._dict['0700'] = "Format Error: invalid type in column 3; must be EXACTLY 'gene' or 'mRNA' or 'exon'."
        
        
class LineError(ValidationError):
    def __init__(self, code, message = ""):
        
        super(LineError, self).__init__(code, message)

        self._dict['0001'] = "Line Error: unknown"
        self._dict['0002'] = "Line Error: 1st component = restrict characters used to a-Z/0-9/./=/;/_"
        self._dict['0003'] = "Line Error: 2nd component = restrict characters used to a-Z/0-9/./=/;/_"
        self._dict['0004'] = "Line Error: 3rd component = type given not found, must exactly match one of [gene, mRNA, exon]."
        self._dict['0005'] = "Line Error: 4th component = 1st coordinate must be a positive integer."
        self._dict['0006'] = "Line Error: 5th component = 2nd coordinate must both an integer and be greater than the 1st coordinate."
        self._dict['0007'] = "Line Error: 6th component = must be a number or '.'"
        self._dict['0008'] = "Line Error: 7th component = must indicate strand as either '+' or '-'; no other characters allowed"
        self._dict['0009'] = "Line Error: 8th component = must be '.'"
        self._dict['0011'] = "Line Error: 9th component = restrict characters used to a-Z/0-9/./=/;/_"
        self._dict['0012'] = "Line Error: 9th component = ID already used. Each gene must have a unique ID."
        self._dict['0013'] = "Line Error: 9th component = Item must have an entry for ID"
        self._dict['0021'] = "Line Error: 9th component = bad sort or a gene does not have a line for each type. Default types are [gene, mRNA, exon]."
        self._dict['0022'] = "Line Error: 9th component = restrict characters used to a-Z/0-9/./=/;/_"
        self._dict['0023'] = "Line Error: 9th component = ID already used. Each ID must be unique."
        self._dict['0024'] = "Line Error: 9th component = last type, must at least have a Parent."
        self._dict['0025'] = "Line Error: 9th component = Lines of this type must have a Parent."
        self._dict['0026'] = "Line Error: First line must be EXACTLY: '##gff-version 3' (HINT: check for unseen spaces or tabs)"
        self._dict['0027'] = "Line Error: 9th component = items in this column must be in the form <name>=<value>, separated by ';'"
        self._dict['0028'] = "Line Error: 9th component = items must be separated by ';', too many '=' without enough ';' "

        
class BiologyError(ValidationError):
    def __init__(self, code, message = ""):
        
        super(BiologyError, self).__init__(code, message)
        
        self._dict['0010'] = "Biology Error: unknown"
        self._dict['0020'] = "Biology Error: Incorrect start codon. Must start with M"
        self._dict['0030'] = "Biology Error: No stop codon detected."
        self._dict['0040'] = "Biology Error: Internal stop codons"
        self._dict['0050'] = "Biology Error: total nucleotide count not divisible by three."
        
```

# Main

```python
def main(gff3_File, seq_File):
    global Errors
    Errors = dict()
    seq = fastaRead(seq_File)
    
    sorted_File = sortGff3(gff3_File)       
    
    report = fileCheck(sorted_File, seq)
    
    newName = gff3_File[0:-5] + "_sorted_Errors.txt"
    out = open(newName,"w")
    for item in Errors:
        out.write(item)
        out.write("\n")
    out.close()
    return [sorted_File, newName]
```

```python
def geneCheckBio(coord1, coord2, seq, count):
    gene = "N/A"
    smlCount = 0
    
    try:
        for theSeq in SeqIO.parse(seq, "fasta"):
            gene = theSeq[coord1:coord2]
            smlCount += 1
        if smlCount != 1:
            raise FormatError("0600")
    except FormatError as er:
        Errors.append("[" + str(count) +"] " + er.returnError())
        return
    except:
        Errors.append("Error in parsing the fasta file. Ensure correct fasta format")
        return
    
    try:
        if len(gene)%3 != 0:
            raise BiologyError("0050") ## divisible by three
    except BiologyError as er:
        Errors.append("[" + str(count) +"] " + er.returnError())
        
    try:
        protein = gene.seq.translate() ## errors in translating the gene
    except:
        Errors.append("[" + str(count) + "]  Biology Error: unable to translate sequence. Ensure sequence provided is a nucleotide sequence.")
        return
    
    try:
        if protein.count("*") == 0:
            raise BiologyError("0030") ## no stop codon
        if protein.count("*") > 1:
            raise BiologyError("0040") ## internal stop codons
        if protein[0] != "M":
            raise BiologyError("0020") ## has to start with M
    except BiologyError as er:
        Errors.append("[" + str(count) +"] " + er.returnError())
    
    return 
```

```python
listKeys = prac.keys()
codeNum = len(listKeys)
codeLen = len(listKeys[0])
prevKey = ""
index = 0
while index < codeNum:
    currKey = listKeys[index]
    if index == 0:
        prevKey = currKey
        index += 1
        continue
    else:
        for y in range(0, codeLen):
            if currKey[y] == prevKey[y]:
                continue
            elif currKey[y] > prevKey[y]:
                listKeys[index-1] = currKey
                listKeys[index] = prevKey
                index -= 2
                break
            else:
                continue
        index += 1
        prevKey = currKey
for key in listKeys:
    print key
        
```
